# Model Validation
Two Techniques: 1. Collinearity Check & 2. Residual Analysis

```python
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_1=smf.ols("RD~Administration+Marketing",data=data).fit().rsquared
vif_1=1/(1-rsq_1)

rsq_2=smf.ols("Administration~RD+Marketing",data=data).fit().rsquared
vif_2=1/(1-rsq_2)

rsq_3=smf.ols("Marketing~RD+Administration",data=data).fit().rsquared
vif_3=1/(1-rsq_3)

# Putting the values in Dataframe format
d1={'Variables':['RD','Administration','Marketing'],'Vif':[vif_1,vif_2,vif_3]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
# No variable has VIF>20, No Collinearity, so consider all varaibles in Regression equation
```

# Model Deletion Diagnostics (checking Outliers or Influencers)
Two Techniques : 1. Cook's Distance & 2. Leverage value

# 1 Cook's distance

```python
(c,_)=model.get_influence().cooks_distance
c
```

```python
# Plot the influencers using the stem plot
fig=plt.figure(figsize=(20,10))
plt.stem(np.arange(len(data)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

```python
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k=data.shape[1]
n=data.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

# improving Data

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
data1=data.drop(data.index[[49]],axis=0).reset_index(drop=True)
data1
```

# Model Deletion Diagnostics and Final Model

```python
model1=smf.ols("Profit~RD+Administration+Marketing",data=data1).fit()
```

```python
while model1.rsquared < 0.99:
    for c in [np.max(c)>1]:
        model1=smf.ols("Profit~RD+Administration+Marketing",data=data1).fit()
        (c,_)=model1.get_influence().cooks_distance
        c
        np.argmax(c) , np.max(c)
        data1=data1.drop(data1.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        data1
    else:
        final_model=smf.ols("Profit~RD+Administration+Marketing",data=data1).fit()
        final_model.rsquared , final_model.aic
        print("Thus model accuracy is improved to",final_model.rsquared)
```
